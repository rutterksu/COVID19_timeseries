```python
# import modules
import pandas as pd # import 'pandas' module and rename to 'pd'
import matplotlib.pyplot as plt # import 'matplotlib.pyplot' and rename to 'plt'
```

```python
# read data
## assign variable names to urls
infected_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv" # url containing data of number of people diagnosed with COVID19 infection
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv" # url containing data of number of people who died from COVID19 infection
recovered_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv" # url containing data of number of people who recovered from COVID19 infection

## scrape data from github
infected_data = pd.read_csv(infected_url) # read data from infected_url and store as dataframe
deaths_data = pd.read_csv(deaths_url) # read data from infected_url and store as dataframe
recovered_data = pd.read_csv(recovered_url) # read data from infected_url and store as dataframe
```

```python
# calculate totals 
total_infected = infected_data.iloc[:,4:].sum() # timeseries of total number of infections
total_deaths = deaths_data.iloc[:,4:].sum() # timeseries of total number of infections
total_recovered = recovered_data.iloc[:,4:].sum() # timeseries of total number of infections
```

```python
# format columns as datetime
dates = pd.to_datetime(infected_data.columns[4:]) # convert date column headings to datetime and save to "dates" variable
```

```python
# create a time-series plot of the COVID19 data
plt.figure(figsize = (8,6))
plt.plot(dates, total_infected, label = "Infected", linestyle = "dashed", color = "darkkhaki") # plot 'total_infected' data as a dashed dark khaki line
plt.plot(dates, total_deaths, label = "Deaths", linestyle = "dotted", color = "red") # plot 'total_deaths' data as a solid green line
plt.plot(dates, total_recovered, label = "Recovered" , linestyle = "solid", color = "green") # plot 'total_recovered' data as a dotted red line
plt.xticks(rotation = 45) # rotate x-axis tick mark labels 45 degrees
plt.legend() # add a legend to the graph using the labels specified above (e.g. "Infected", "Deaths", "Recovered")
plt.ylabel("People\n") # Specify the vertical axis title
plt.title("Coronavirus Timeseries") # Specify the figure title
plt.show() # print the plot
```
